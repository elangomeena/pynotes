```python
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Division by zero error"
    except TypeError:
        return "Invalid input type"

print(divide(10, 5))
```

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Handled division by zero safely")
```
