```python
def validate_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")

validate_age(-5)
```

```python
class NegativeNumberError(Exception):
    pass

def check_number(n):
    if n < 0:
        raise NegativeNumberError("Negative numbers are not allowed")

check_number(-3)
```
