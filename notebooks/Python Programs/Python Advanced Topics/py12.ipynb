```python
def resource_handler():
    try:
        yield "Using resource"
    finally:
        print("Resource released")

gen = resource_handler()
print(next(gen))
```

```python
def read_large_file(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()

for line in read_large_file("data.txt"):
    print(line)
```
