```python
def log_decorator(func):
    def wrapper(*args, **kwargs):
        print("Arguments:", args, kwargs)
        return func(*args, **kwargs)
    return wrapper

@log_decorator
def add(a, b):
    return a + b

print(add(5, 3))
```

```python
def bold(func):
    def wrapper():
        print("<b>")
        func()
        print("</b>")
    return wrapper

def italic(func):
    def wrapper():
        print("<i>")
        func()
        print("</i>")
    return wrapper

@bold
@italic
def display():
    print("Text")

display()
```

```python
def uppercase(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper

@uppercase
def message():
    return "hello world"

print(message())
```

```python
from functools import wraps

def my_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper
```
