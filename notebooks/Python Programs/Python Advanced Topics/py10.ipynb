```python
def simple_generator():
    yield 1
    yield 2
    yield 3

gen = simple_generator()
print(next(gen))  # 1
print(next(gen))  # 2
```

```python
def normal_function():
    return 10
    return 20

def generator_function():
    yield 10
    yield 20

print(normal_function())          # 10
print(list(generator_function())) # [10, 20]
```

```python
def count_up(n):
    for i in range(1, n + 1):
        yield i

for num in count_up(5):
    print(num)
```

```python
def demo():
    print("Start")
    yield 1
    print("Resume")
    yield 2

g = demo()
print(next(g))
print(next(g))
```
