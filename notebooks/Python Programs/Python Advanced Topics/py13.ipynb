```python
x = "Global"

def outer():
    x = "Enclosing"
    def inner():
        x = "Local"
        print(x)
    inner()

outer()
```

```python
def outer():
    message = "Hello"
    def inner():
        print(message)
    inner()

outer()
```

```python
def counter():
    value = 0
    def increment():
        nonlocal value
        value += 1
        return value
    return increment

count = counter()
print(count())
print(count())
```

```python
x = "Global"

def test():
    x = "Local"
    print("Inside:", x)

test()
print("Outside:", x)
```
