```python
def outer():
    message = "Hello"
    
    def inner():
        return message

    return inner

func = outer()
print(func())  # Output: Hello
```

```python
def multiplier(x):
    def multiply(y):
        return x * y
    return multiply

times2 = multiplier(2)
print(times2(5))  # Output: 10
```

```python
def counter():
    count = 0

    def increment():
        nonlocal count
        count += 1
        return count

    return increment

c = counter()
print(c())  # 1
print(c())  # 2
```

```python
def power(base):
    exponent = 2

    def calculate():
        nonlocal exponent
        exponent += 1
        return base ** exponent

    return calculate

p = power(2)
print(p())
print(p())
```
