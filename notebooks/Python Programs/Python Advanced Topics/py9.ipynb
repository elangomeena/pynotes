```python
class InfiniteNumbers:
    def __iter__(self):
        self.num = 0
        return self

    def __next__(self):
        self.num += 1
        return self.num
```

```python
data = "Python"

it = iter(data)
print(list(it))  # ['P', 'y', 't', 'h', 'o', 'n']
```

```python
from collections.abc import Iterator

numbers = iter([1, 2, 3])
print(isinstance(numbers, Iterator))  # True
```

```python
class FileLineReader:
    def __init__(self, filename):
        self.file = open(filename, "r")

    def __iter__(self):
        return self

    def __next__(self):
        line = self.file.readline()
        if line:
            return line.strip()
        self.file.close()
        raise StopIteration

for line in FileLineReader("data.txt"):
    print(line)
```
