```python
factor = 10

def using_global():
    return lambda x: x * factor

def using_closure():
    factor = 10
    return lambda x: x * factor

print(using_closure()(5))
```

```python
def make_adder(x):
    def adder(y):
        return x + y
    return adder

add5 = make_adder(5)
add10 = make_adder(10)

print(add5(3))    # 8
print(add10(3))   # 13
```

```python
def outer():
    value = 42

    def inner():
        return value

    return inner

func = outer()
print(func.__closure__[0].cell_contents)  # 42
```

```python
def logger(prefix):
    def log(message):
        print(f"{prefix}: {message}")
    return log

info_logger = logger("INFO")
error_logger = logger("ERROR")

info_logger("System started")
error_logger("System failed")
```
