```python
def large_numbers():
    for i in range(1000000):
        yield i

gen = large_numbers()
print(next(gen))
print(next(gen))
```

```python
squares = (x ** 2 for x in range(5))
print(list(squares))  # [0, 1, 4, 9, 16]
```

```python
def alpha_generator():
    yield "A"
    yield "B"
    yield "C"

gen = alpha_generator()
print(next(gen))
print(next(gen))
```

```python
def infinite_counter():
    num = 1
    while True:
        yield num
        num += 1

counter = infinite_counter()
print(next(counter))
print(next(counter))
```
