```python
import time

start = time.time()
time.sleep(1)
end = time.time()

print("Execution Time:", end - start, "seconds")
```

```python
import time

start = time.perf_counter()
time.sleep(0.5)
end = time.perf_counter()

print("Precise Time:", end - start)
```

```python
import time

formatted = time.strftime("%Y-%m-%d %H:%M:%S")
print(formatted)
```

```python
import time

parsed = time.strptime("24-11-2025 10:30:00", "%d-%m-%Y %H:%M:%S")
print(parsed)
```
