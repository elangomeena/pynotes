```python
import time

timestamp = time.time()
readable = time.ctime(timestamp)

print(readable)
```

```python
import time

last_request = time.time()

def can_execute():
    global last_request
    current = time.time()
    if current - last_request >= 5:
        last_request = current
        return True
    return False

print(can_execute())
```
