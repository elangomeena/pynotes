```python
import json

def parse_api_response(response):
    try:
        data = json.loads(response)
        return data.get("status"), data.get("payload")
    except json.JSONDecodeError:
        return "error", None

response = '{"status": "success", "payload": {"id": 101}}'
print(parse_api_response(response))
```
