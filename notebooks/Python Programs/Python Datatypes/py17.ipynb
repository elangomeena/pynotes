```python
A = {1, 2, 3}
B = {1, 2}

print(B.issubset(A))     # True
print(A.issuperset(B))   # True
```

```python
values_set = {1, 2, 3}

list_version = list(values_set)
tuple_version = tuple(values_set)

print(list_version)   # [1, 2, 3]
print(tuple_version)  # (1, 2, 3)
```
