```python
numbers = [0, 1, 2, 3, 4, 5]

print(numbers[1:4])   # Output: [1, 2, 3]
print(numbers[:3])    # Output: [0, 1, 2]
print(numbers[3:])    # Output: [3, 4, 5]
```

```python
items = ["pen", "pencil", "eraser"]
items[1] = "marker"

print(items)  # Output: ['pen', 'marker', 'eraser']
```

```python
numbers = [1, 2, 3]

numbers.append(4)
numbers.insert(1, 10)

print(numbers)  # Output: [1, 10, 2, 3, 4]
```
