```python
class A:
    def __init__(self):
        print("A init")
        super().__init__()

class B:
    def __init__(self):
        print("B init")
        super().__init__()

class C(A, B):
    def __init__(self):
        print("C init")
        super().__init__()

c = C()
```

```python
class Grandparent:
    def greet(self):
        print("Hello from Grandparent")

class Parent1(Grandparent):
    pass

class Parent2(Grandparent):
    pass

class Child(Parent1, Parent2):
    pass

c = Child()
c.greet()
```

```python
class Logger:
    def log(self):
        print("Logging data")

class Validator:
    def validate(self):
        print("Validating data")

class Service(Logger, Validator):
    def execute(self):
        self.log()
        self.validate()
        print("Executing service")

service = Service()
service.execute()
```
