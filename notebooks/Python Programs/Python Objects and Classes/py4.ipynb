```python
class Animal:
    def eat(self):
        print("Animal is eating")

class Dog(Animal):
    pass

d = Dog()
d.eat()  # Inherited method
```

```python
class Parent:
    def show(self):
        print("This is the parent class")

class Child(Parent):
    pass

c = Child()
c.show()
```

```python
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.sound()
d.bark()
```
