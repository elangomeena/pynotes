```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, my name is {self.name}"

p = Person("Alice")
print(p.greet())
```

```python
class Student:
    school = "Global Academy"  # Class variable

    def __init__(self, name):
        self.name = name        # Instance variable

s1 = Student("Alice")
s2 = Student("Bob")

print(s1.school)
print(s2.school)
```

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

c1 = Car("Toyota")
c2 = Car("Tesla")

print(c1.brand)
print(c2.brand)
```
