```python
x = 20  # Global variable

def display():
    print(x)

display()
print(x)
```

```python
value = 50

def update():
    value = 10  # Local variable shadows global
    print("Inside function:", value)

update()
print("Outside function:", value)
```

```python
count = 0

def increment():
    global count
    count += 1

increment()
print(count)  # Output: 1
```
