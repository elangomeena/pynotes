```python
def print_list(lst):
    if not lst:
        return
    print(lst[0])
    print_list(lst[1:])

print_list([1, 2, 3, 4])
```

```python
def reverse_string(text):
    if len(text) == 0:
        return text
    return reverse_string(text[1:]) + text[0]

print(reverse_string("Python"))  # Output: nohtyP
```

```python
def infinite_recursion():
    return infinite_recursion()

infinite_recursion()  # Raises RecursionError
```
