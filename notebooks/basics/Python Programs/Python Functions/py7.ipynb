```python
def configure(*,mode="light"):
    print(f"Mode: {mode}")

configure(mode="dark")
```

```python
def multiply(a, b, c):
    return a * b * c

values = (2, 3, 4)
print(multiply(*values))  # Output: 24

data = {"a": 1, "b": 2, "c": 3}
print(multiply(**data))   # Output: 6
```
