```python
requests = 0

def handle_request():
    global requests
    requests += 1

handle_request()
handle_request()
print(requests)  # Output: 2
```

```python
mode = "light"

def toggle_mode():
    global mode
    if mode == "light":
        mode = "dark"
    else:
        mode = "light"

toggle_mode()
print(mode)  # Output: dark
```

```python
config = {"theme": "dark"}

def update_config(new_theme):
    config["theme"] = new_theme
    return config

print(update_config("light"))
```
